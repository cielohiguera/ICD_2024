##### _Las abejas melíferas, tanto silvestres como domésticas, realizan el 80 por ciento de toda la polinización en el mundo. Una sola colonia de abejas puede polinizar 300 millones de flores cada día._ 

##### _Actualmente, existe un colapso de colonias de abejas en todo el mundo. Existen diversos factores por los cuales las abejas están muriendo, algunos ejemplos son los pesticidas, sequía, destrucción de hábitat, plagas, déficit nutricional, contaminación del aire, calentamiento global, entre otros._ 

##### _Con la disminución de las poblaciones de abejas, ha habido un aumento de la investigación centrada en los factores que impulsan la disminución de las abejas y los impactos en el aprovisionamiento de servicios ecosistémicos (Patel, V., 2021)._ 

##### _En Estados Unidos, las pérdidas invernales han llegado a ser del 30 al 50 por ciento, en algunos casos más. Las  estadísticas agrícolas nacionales de Estados Unidos  muestran una disminución de la población de abejas melíferas de unos 6 millones de colmenas en 1947 a 2,4 millones en 2008, una reducción del 60 por ciento (Andrews, E., 2022)._

###### Acerca de las columnas:  

###### _state:_ Estado dentro de los EE. UU.

###### _num_colonies:_ Número de colonias de abejas melíferas

###### _max_colonies:_ Número máximo de colonias de abejas melíferas para ese trimestre.

###### _lost_colonies:_ Número de colonias que se perdieron durante ese trimestre.

###### _percent_lost:_ Porcentaje de colonias de abejas perdidas durante ese trimestre

###### _renovated_colonies:_ colonias que fueron reubicadas o recibieron nuevas abejas

###### _percent_renovated:_ Porcentaje de colonias de abejas melíferas que fueron renovadas

###### _quarter:_ Q1es de enero a marzo, Q2de abril a junio, Q3de julio a septiembre y Q4de octubre a diciembre

###### _year:_ Año entre 2015 y 2022

###### _varroa_mites:_ Porcentaje de colonias afectadas por una especie de ácaro que afecta a las poblaciones de abejas melíferas

###### _other_pests_and_parasites:_ Porcentaje de colonias afectadas por una colección de otras criaturas dañinas

###### _diseases:_ Porcentaje de colonias afectadas por determinadas enfermedades

###### _pesticides:_ Porcentaje de colonias afectadas por el uso de determinados plaguicidas

###### _other:_ Porcentaje de colonias afectadas por una causa no listada

###### _unknown:_ Porcentaje de colonias afectadas por una causa desconocida

#### Agrupaciones de datos

```python
group_year = df.groupby('year')['num_colonies'].sum().reset_index()
highest_colonies = group_year.sort_values(by='num_colonies',ascending=False)
highest_colonies.reset_index(drop=True, inplace=True)
print('Año con mayor cantidad de colonias:')
print(highest_colonies.head(10))
##
group_year = df.groupby('year')['num_colonies'].sum().reset_index()
lowest_colonies = group_year.sort_values(by='num_colonies',ascending=True)
lowest_colonies.reset_index(drop=True, inplace=True)
print('Año con menor cantidad de colmenas:')
print(lowest_colonies.head(10))
```

##### De acuerdo a los datos, se considera que en el año 2020 se presentaron mayores cantidad de colmenas en Estados Unidos, con un número de colmenas de 12,158,770 colmenas. Así mismo, el año 2019 se presentaron menos cantidad de colmenas en Estados Unidos con un número de colmenas de 8,865,540 colmenas.

```python
group_year = df.groupby('year')['added_colonies'].sum().reset_index()
more_added_colonies = group_year.sort_values(by='added_colonies',ascending=False)
more_added_colonies.reset_index(drop=True, inplace=True)
print('Año con mayor cantidad de colmenas añadidas:')
print(more_added_colonies.head(10))
#
group_year = df.groupby('year')['added_colonies'].sum().reset_index()
fewer_added_colonies = group_year.sort_values(by='added_colonies',ascending=True)
fewer_added_colonies.reset_index(drop=True, inplace=True)
print('Año con menor cantidad de colmenas añadidas:')
print(fewer_added_colonies.head(10))
```

##### De acuerdo a los datos, se considera que en el año 2018 se presentó mayor cantidad de colmenas añadidas en Estados Unidos, con un número de colmenas de 1,670,880 colmenas. Así mismo, el año 2019 se presentó menor cantidad de colmenas añadidas en Estados Unidos con un número de colmenas de 7,335,520 colmenas.

```python
group_year = df.groupby('year')['lost_colonies'].sum().reset_index()
lostest_colonies = group_year.sort_values(by='lost_colonies',ascending=False)
lostest_colonies.reset_index(drop=True, inplace=True)
print('Año con mayor cantidad de colmenas perdidas:')
print(lostest_colonies.head(10))
#
group_year = df.groupby('year')['lost_colonies'].sum().reset_index()
fewer_lost_colonies = group_year.sort_values(by='lost_colonies',ascending=True)
fewer_lost_colonies.reset_index(drop=True, inplace=True)
print('Año con menor cantidad de colmenas perdidas:')
print(fewer_lost_colonies.head(10))
```

##### De acuerdo a los datos, se considera que en el año 2015 se presentó mayor cantidad de colmenas perdidas en Estados Unidos, con un número de colmenas de 1,722,360 colmenas. Así mismo, el año 2019 se presentó menor perdidas de colmenas añadidas en Estados Unidos con un número de colmenas de 1,241,910 colmenas.

```python
grouped = df.groupby('year')['percent_lost'].apply(list)
years = df['year'].unique()  
fig, ax = plt.subplots(figsize=(10, 8))
palette = sns.color_palette("pastel")
box = ax.boxplot(grouped, labels=years, patch_artist=True)
for patch, color in zip(box['boxes'], palette):
    patch.set_facecolor(color)
for median in box['medians']:
    median.set_color('black')
    
ax.set_title('The Percentage Lost Per Year (2015-2022)')
ax.set_xlabel('Year')
ax.set_ylabel('Percentage Lost')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))  
bars = ax.bar(df['state'], df['num_colonies'], color='skyblue', edgecolor='none')
ax.set_title('State with Total Number of Colonies Per Year (2015-2022)', fontsize=14)
ax.set_xlabel('State', fontsize=12)
ax.set_ylabel('Number of Colonies', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)  
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 
plt.show()
```

##### En el siguiente gráfico se puede apreciar que California es el estado con mayor número de colonias agregadas en el periodo de 2015 a 2022. Los siguientes estados con mayor perdidas de colmenas es Texas, Florida y North Dakota.

```python
if 'pesticides' in df.columns and 'other_pests_and_parasites' in df.columns and 'varroa_mites' in df.columns and 'diseases' in df.columns and 'other':
    promedios = df[['pesticides', 'other_pests_and_parasites', 'varroa_mites', 'diseases', 'other']].mean()
    
    plt.figure(figsize=(8, 8))
    plt.pie(promedios, labels=promedios.index, autopct='%1.1f%%', startangle=140)
    plt.title('Pirncipales causas de perdidas de colmenas en Estados Unidos')
    plt.show()
```

##### A partir de un análisis de las principales causas de pérdidas de colmenas de Estados Unidos, se puede concluir que el porcentaje de colonias afectadas por una especie de ácaro (53.2%) afecta con más recurrencia las colmenas, sin embargo, algunas causas comunes que las colmenas puede ser animales y parásitos (19.3%), pesticidas (10.8%), otros (10.7%) y enfermedades (6.0%).

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['pesticides'], df['lost_colonies'], alpha=0.7, edgecolors='w', s=100)
ax.set_title('Relationship Between Pesticide Usage and Lost Colonies')
ax.set_xlabel('Pesticide Usage')
ax.set_ylabel('Lost Colonies')
plt.tight_layout()
plt.show()
```

##### A partir del gráfico correspondiente se puede observar que el pesticida tiene relación con las colonias perdidas.
