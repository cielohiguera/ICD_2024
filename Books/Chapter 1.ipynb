##### Profiling Memory Use: %Memit and %mprun

```python
%load_ext memory_profiler
```

```python
%memit sum_of_lists(1000000)
```

```python
%%file mprun_demo.py
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
        del L # remove reference to L
return total
```
